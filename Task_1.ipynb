**Detecting and Handling Outliers**

```python

numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns

outlier_summary = {}

for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_limit = Q1 - 1.5 * IQR
    upper_limit = Q3 + 1.5 * IQR


    outliers = df[(df[col] < lower_limit) | (df[col] > upper_limit)]
    outlier_summary[col] = outliers.shape[0]

outlier_df = pd.DataFrame.from_dict(outlier_summary, orient='index', columns=['Outlier_Count'])
outlier_df = outlier_df.sort_values(by='Outlier_Count', ascending=False)

print("Outlier Summary across all numerical columns:")
display(outlier_df)
```

**detect outliers using Z-score method**

```python
from scipy import stats
import numpy as np
import pandas as pd

numeric_cols = df.select_dtypes(include=['int64', 'float64'])

z_scores = np.abs(stats.zscore(numeric_cols))

outlier_mask = (z_scores > 3)

outlier_counts = outlier_mask.sum(axis=0)

outlier_summary = pd.DataFrame({
    'Column': numeric_cols.columns,
    'Outlier_Count': outlier_counts
}).sort_values(by='Outlier_Count', ascending=False)

print("Outlier Summary using Z-Score:")
display(outlier_summary)

total_outliers = (outlier_mask.sum().sum())
print(f"\nTotal Outliers in Dataset: {total_outliers}")
```

**cap outliers (Winsorizing)**

```python
def cap_outliers(data, column, lower_quantile=0.05, upper_quantile=0.95):
    lower_bound = data[column].quantile(lower_quantile)
    upper_bound = data[column].quantile(upper_quantile)

    data_capped = data.copy()
    data_capped[column] = np.where(data_capped[column] < lower_bound, lower_bound,
                                 np.where(data_capped[column] > upper_bound, upper_bound,
                                         data_capped[column]))
    return data_capped
```

<h3><b>Column Name Standardization (snake_case)</b></h3>

```python
def to_snake_case(name):
    new_name = ""
    for i, ch in enumerate(name):
        if i > 0 and ch.isupper() and not name[i-1].isspace():
            new_name += "_"
        new_name += ch.lower()
    return new_name.replace(" ", "_")

df.columns = [to_snake_case(col) for col in df.columns]
df.columns
```

<h3><b>Removing Outliers From monthly_income column</b></h3>

```python
def remove_outliers_monthly_income(df):
    clean_df = df.copy()
    col = 'monthly_income'

    if col not in clean_df.columns:
        raise ValueError(f"Column '{col}' not found in DataFrame.")

    changed = True
    while changed:
        changed = False
        initial_shape = clean_df.shape[0]

        Q1 = clean_df[col].quantile(0.25)
        Q3 = clean_df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        clean_df = clean_df[(clean_df[col] >= lower_bound) & (clean_df[col] <= upper_bound)]

        if clean_df.shape[0] < initial_shape:
            changed = True

    return clean_df.reset_index(drop=True)

clean_df = remove_outliers_monthly_income(df)
print("Original shape:", df.shape)
print("After IQR outlier removal:", clean_df.shape)
```
